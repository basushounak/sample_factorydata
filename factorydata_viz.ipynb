```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the .csv dataset into Pandas DataFrame

df = pd.read_csv('millingmachinedata1.csv', encoding='ISO-8859-1')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python

sns.scatterplot(data=df, x='Power Consumption (kW)', y='Temperature')

#Setting Lables and Title 
plt.xlabel('Power Consumption')
plt.ylabel('Temperature')
plt.title("kW vs T Scatter Plot")

plt.show()
```
